#4. Computing NN accuracy using SFrame operations

```python
#splitting test data into one SFrame for each category
image_test_cat = test_data[test_data['label'] == 'cat']
image_test_dog = test_data[test_data['label'] == 'dog']
image_test_bird = test_data[test_data['label'] == 'bird']
image_test_automobile = test_data[test_data['label'] == 'automobile']
```

```python
#finding nearest neighbors in the training set for each part of the test set
cat_model.query(image_test_cat)
```

```python
#find the closest 1 neighbor in to the dog test data using each of the trained models
dog_cat_neighbors = cat_model.query(image_test_dog, k=1)
```

```python
dog_automobile_neighbors = automobile_model.query(image_test_dog, k=1)
dog_bird_neighbors = bird_model.query(image_test_dog, k=1)
dog_dog_neighbors = dog_model.query(image_test_dog, k=1)
```

```python
dog_cat_neighbors
```

## Create an SFrame with the distances from 'dog' test examples to the respective nearest neighbors in each category in training data

```python
#create an SFrame called dog_distances with 4 columns:
dog_distances = graphlab.SFrame({'dog_automobile': dog_automobile_neighbors['distance'],
                                 'dog_bird': dog_bird_neighbors['distance'],
                                 'dog_cat': dog_cat_neighbors['distance'], 
                                 'dog_dog': dog_dog_neighbors['distance']})
```

```python
dog_distances
```

```python
#consider one row of the dog_distance SFrame, and call this variable row
row = dog_distances[0:1]
row['dog_cat']
```

```python
def is_dog_correct(row):
    if row['dog_cat'] < row['dog_dog']
    return 1
```
