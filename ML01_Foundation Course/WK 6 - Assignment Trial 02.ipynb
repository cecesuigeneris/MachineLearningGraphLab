#2. Category-specific image retrieval models

```python
dog_filter = image_train[image_train['label'] == 'dog']
```

```python
dog_filter.head()
```

```python
cat_filter = image_train[image_train['label'] == 'cat']
```

```python
automobile_filter = image_train[image_train['label'] == 'automobile']
```

```python
bird_filter = image_train[image_train['label'] == 'bird']
```

```python
cat_model = graphlab.nearest_neighbors.create(cat_filter,
                                             features=['deep_features'],
                                             label='id') 
```

```python
cat = image_test[0:1]
```

```python
cat['image'].show()
```

```python
cat_model.query(cat)
```

```python
dog_model = graphlab.nearest_neighbors.create(dog_filter,
                                             features=['deep_features'],
                                             label='id') 
```

```python
dog_model.query(cat)
```

```python
def get_cat_ids(query_result):
    return cat_filter.filter_by(query_result['reference_label'],'id')
```

```python
cat_neighbors = get_cat_ids(cat_model.query(cat))
```

```python
cat_neighbors.head()
```

```python
cat_neighbors['image'].show()
```

```python
def get_dog_ids(query_result):
    return dog_filter.filter_by(query_result['reference_label'],'id')
```

```python
dog_neighbors = get_dog_ids(dog_model.query(cat))
```

```python
dog_neighbors['image'].show()
```

#4. Computing nearest neighbors accuracy using SFrame operations

```python
#splitting test data into one SFrame for each category
image_test_cat = image_test[image_test['label'] == 'cat']
image_test_dog = image_test[image_test['label'] == 'dog']
image_test_bird = image_test[image_test['label'] == 'bird']
image_test_automobile = image_test[image_test['label'] == 'automobile']
```

```python
def get_dogtest_ids(query_result):
    return image_test_dog.filter_by(query_result['reference_label'],'id')
```

```python
dog_cat_neighbors = get_dogtest_ids(cat_model.query(image_test_dog, k=1))
```

```python
dog_cat_neighbors['image'].show()
```

```python
bird_model = graphlab.nearest_neighbors.create(bird_filter,
                                             features=['deep_features'],
                                             label='id') 
```

```python
automobile_model = graphlab.nearest_neighbors.create(automobile_filter,
                                             features=['deep_features'],
                                             label='id') 
```

```python
dog_cat_neighbors.head()
```

```python
dog_dog_neighbors = get_dogtest_ids(dog_model.query(image_test_dog, k=1))
dog_bird_neighbors = get_dogtest_ids(bird_model.query(image_test_dog, k=1))
dog_automobile_neighbors = get_dogtest_ids(automobile_model.query(image_test_dog, k=1))
```

```python
#create an SFrame called dog_distances with 4 columns:
dog_distances = graphlab.SFrame({'dog_automobile': dog_automobile_neighbors['distance'],
                                 'dog_bird': dog_bird_neighbors['distance'],
                                 'dog_cat': dog_cat_neighbors['distance'], 
                                 'dog_dog': dog_dog_neighbors['distance']})
```
